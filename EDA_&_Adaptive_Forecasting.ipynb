# Adaptive Forecasting Models

First, we get the shape for quarterly and monthly sales to split into training and testing sets.

```python
quartly_sales.shape
```

```python
monthly_sales.shape
```

Now, we want to split the time series data for training and testing sets for quarterly and monthly sales.

Since monthly sales has 48 datasets, 42 months are being split into training data and 6 months is split into testing data,which will be used for forecasting.

The same goes to quarterly sales where 12 periods are being split into training set and 4 periods are testing set.

```python
#split quarterly sales time series data

q_train_len = 12
q_train = quartly_sales[0:q_train_len]
q_test = quartly_sales [q_train_len:]
```

```python
#split monthly sales time series data

m_train_len = 42
m_train = monthly_sales[0:m_train_len]
m_test = monthly_sales [m_train_len:]
```

### Simple Moving Average

```python
#Simple Moving Average Time Series Forecasting for Monthly Sales

y_m_sma = monthly_sales.copy()
ma_window = 3  # Considered 3 month window
y_m_sma['sma_forecast'] = monthly_sales['Sales'].rolling(ma_window).mean()
y_m_sma['sma_forecast'][m_train_len:] = y_m_sma['sma_forecast'][m_train_len-1]
```

```python
y_m_sma.head()
```

```python
#Plot graph of Monthly Simple Moving Average Method
plt.figure(figsize=(12,4))
plt.plot(m_train['Sales'], label='Train')
plt.plot(m_test['Sales'], label='Test')
plt.plot(y_m_sma['sma_forecast'], label='Simple moving average forecast')
plt.legend(loc='best')
plt.title('Simple Moving Average Method for Monthly Sales', fontweight= 'bold')
#plt.savefig('Simple Moving Average Method for Monthly Sales.png')
plt.grid(True)
plt.show()
```

We then calculate the forecasting error by finding RMSE and MAPE with monthly sales

```python
rmse = np.sqrt(mean_squared_error(m_test['Sales'], y_m_sma['sma_forecast'][m_train_len:])).round(2)
mape = np.round(np.mean(np.abs(m_test['Sales']-y_m_sma['sma_forecast'][m_train_len:])/m_test['Sales'])*100,2)

#plot the results into a dataframe
sma_m_results = pd.DataFrame({'Method':['Simple moving average forecast monthly'], 'RMSE': [rmse],'MAPE': [mape] })
results = sma_m_results[['Method', 'RMSE', 'MAPE']]
results
```

Now, let's repeat the same steps for quarterly sales

```python
#Simple Moving Average Time Series Forecasting for Quarterly Sales

y_q_sma = quartly_sales.copy()
ma_window = 3  # Considered 3 month window
y_q_sma['sma_forecast'] = quartly_sales['Sales'].rolling(ma_window).mean()
y_q_sma['sma_forecast'][q_train_len:] = y_q_sma['sma_forecast'][q_train_len-1]
```

```python
y_q_sma.head()
```

```python
#Plot graph of Quarterly Simple Moving Average Method
plt.figure(figsize=(12,4))
plt.plot(q_train['Sales'], label='Train')
plt.plot(q_test['Sales'], label='Test')
plt.plot(y_q_sma['sma_forecast'], label='Simple moving average forecast')
plt.legend(loc='best')
plt.title('Simple Moving Average Method for Quarterly Sales', fontweight= 'bold')
#plt.savefig('Simple Moving Average Method for Quarterly Sales.png')
plt.grid(True)
plt.show()
```

```python
#Find Forecasting error for quarterly sales with SMA method

rmse = np.sqrt(mean_squared_error(q_test['Sales'], y_q_sma['sma_forecast'][q_train_len:])).round(2)
mape = np.round(np.mean(np.abs(q_test['Sales']-y_q_sma['sma_forecast'][q_train_len:])/q_test['Sales'])*100,2)

#plot the results into a dataframe
sma_q_results = pd.DataFrame({'Method':['Simple moving average forecast quarterly'], 'RMSE': [rmse],'MAPE': [mape] })
results = pd.concat([sma_m_results, sma_q_results])
results
```

### Simple Exponential Smoothing

First, let's perform simple exponential smoothing for monthly sales.

```python
m_model = SimpleExpSmoothing(m_train['Sales'])
#model_fit = model.fit(smoothing_level = 0.2, optimized=True)
m_model_fit = m_model.fit(optimized=True)
m_model_fit.params
y_m_ses = m_test.copy()
y_m_ses['ses_forecast'] = m_model_fit.forecast(steps = 12)
```

```python
y_m_ses.head()
```

```python
#Plot graph of Monthly Simple Exponential Method
plt.figure(figsize=(12,4))
plt.plot(m_train['Sales'], label='Train')
plt.plot(m_test['Sales'], label='Test')
plt.plot(y_m_ses['ses_forecast'], label='Simple exponential smoothing forecast')
plt.legend(loc='best')
plt.title('Simple Exponential Smoothing Method for Monthly Sales')
plt.show()
```

```python
#Calculate error for monthly sales using simple exponential smoothing method

rmse = np.sqrt(mean_squared_error(m_test['Sales'], y_m_ses['ses_forecast'])).round(2)
mape = np.round(np.mean(np.abs(m_test['Sales']-y_m_ses['ses_forecast'])/m_test['Sales'])*100,2)

ses_m_results = pd.DataFrame({'Method':['Simple exponential smoothing forecast monthly'], 'RMSE': [rmse],'MAPE': [mape] })
results = pd.concat([sma_m_results,sma_q_results,ses_m_results])
results
```

Now we repeat for quarterly sales.

```python
q_model = SimpleExpSmoothing(q_train['Sales'])
#model_fit = model.fit(smoothing_level = 0.2, optimized=True)
q_model_fit = q_model.fit(optimized=True)
q_model_fit.params
y_q_ses = q_test.copy()
y_q_ses['ses_forecast'] = q_model_fit.forecast(steps = 12)
```
